```python
import os
import shutil
check_path=r'E:\python_xml_creation\Tensorflow_data'
output_path=r'E:\python_xml_creation\original_data'
final_count=0
for folders in os.listdir(check_path):
    count=0
    print('Finding original images for {} anomaly....'.format(folders))
    image_list=[]
    [image_list.append(i) for i in os.listdir(os.path.join(check_path,folders)) if ((i.endswith('.')) and 
                                                                                    (i.startswith(tuple(['co','sh']))==False))]
    if not(os.path.exists(os.path.join(output_path,folders))):
        os.mkdir(os.path.join(output_path,folders))
    for j in image_list:
        shutil.copy(os.path.join(os.path.join(check_path,folders),j),os.path.join(os.path.join(output_path,folders),j))
        count+=1
    print('{} files found and written to output directory'.format(count))
    final_count+=count
print('Original Images count :{}'.format(final_count))
        
```
